```python
def mae(y, ypred):
  yDiff = ypred - y.reshape(-1, 1)
  yAbs = abs(yDiff)
  erroSum = sum(yAbs)

  return erroSum/len(yAbs)
```

```python
print(mae(y, ypred))
```

```python
def mse(y, ypred):
  yDiff = ypred - y.reshape(-1, 1)
  yDiffSquare = yDiff**2
  erroSum = sum(yDiffSquare)

  return erroSum/len(ypred)
```

```python
print(mse(y, ypred))
```

```python
def rmse(y, ypred):
  mse = mean_squared_error(y, ypred)

  return sqrt(mse)
```

```python
import numpy as np

class regressao_linear():
    def __init__(self):
        self.w = np.zeros((2, 1))

    @staticmethod
    def includeBias(X):
        bias = np.ones((X.shape[0], 1))
        Xb = np.concatenate((bias, X), axis=1)
        return Xb

    def fit(self, X, y):
        Xb = regressao_linear.includeBias(X)
      
        for _ in range(5000):
            ypred = self.predict(X)
            erros = ypred - y.reshape(-1, 1)
            ajuste = Xb.T.dot(erros)/np.sum(Xb,axis=0)#Xb.T.dot(erros)/sum(Xb)
            self.w = self.w - ajuste*0.1

    def predict(self, X, y=None):
        Xb = regressao_linear.includeBias(X)
        ypred = Xb.dot(self.w)
        return ypred
```

```python
modelo = regressao_linear()
modelo.fit(X12, y)
ypred = modelo.predict(X12)

plt.scatter(X12, y)
plt.plot(X12, ypred)
```
